При использовании алгоритма Bagging каждая модель не видела ту или иную часть датасета, поскольку мы делаем выборку с возвращением. 
И эта часть равна примерно 37% от датасета, поскольку вероятность не попасть в выборку для каждого отдельного объекта:
$$
\lim_{ N \to \infty } {\left( 1-\frac{1}{N} \right)^N}
$$
Данный предел равен:
$$
\frac{1}{e}\approx {0.368}
$$
В этом ноутбуке я продемонстрирую точность данной оценки.

**Стоит заметить, что при Bagging создаваемые выборки должны быть равны по размеру изначальному датасету**

```python
import pandas as pd

orig_dset = pd.read_csv('/kaggle/input/cifar10-python-in-csv/train.csv')
```

```python
import random
def check_lim(amount, dataset):
    test_dset = dataset.iloc[:amount]
    sample = test_dset.iloc[random.choices(range(amount), k=amount)]
    print(100 * (amount - len(set(sample.index)))/amount, "% наблюдений не попало в датасет, количество наблюдений: ", amount)

check_lim(amount=1000, dataset=orig_dset)
```

```python
import random
import numpy as np
observations = []
def check_lim(amount, dataset):
    test_dset = dataset.iloc[:amount]
    sample = test_dset.iloc[random.choices(range(amount), k=amount)]
    observations.append(100 * (amount - len(set(sample.index)))/amount)

for i in range(1000):
    check_lim(amount=10000, dataset=orig_dset)

print(np.mean(observations), "% не попадало в датасет")
```
